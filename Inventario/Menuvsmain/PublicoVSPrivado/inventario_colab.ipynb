```python

#cruz barron
class Producto:
    def __init__(self, nombre, cantidad, precio):
        # ATRIBUTOS PRIVADOS
        self.__nombre = nombre
        self.__cantidad = cantidad
        self.__precio = precio

    def __str__(self):
        return f"{self.__nombre} | Stock: {self.__cantidad} | Precio: ${self.__precio}"

    def agregar_stock(self, cantidad):
        self.__cantidad += cantidad

    def vender(self, cantidad):
        self.__cantidad -= cantidad

    def get_nombre(self):
        return self.__nombre

    def get_cantidad(self):
        return self.__cantidad

    def get_precio(self):
        return self.__precio
```

```python
# clase Departament
class Departamento:
    def __init__(self, nombre_depa, nombre_prod, cant_prod, precio_prod):

        self.__nombre_depa = nombre_depa
        self.__producto_asignado = Producto(nombre_prod, cant_prod, precio_prod)

    def __str__(self):
        return f"Departamento: {self.__nombre_depa.upper()} -> Inventario: [{self.__producto_asignado}]"


    def agregar_stock(self, cantidad):
        self.__producto_asignado.agregar_stock(cantidad)

    def reducir_stock(self, cantidad):
        self.__producto_asignado.vender(cantidad)

    def get_producto(self):
        return self.__producto_asignado
```

```python
# 3. Validaciones
def validar_ingreso(cantidad):
    """Verifica que la cantidad a ingresar sea un número positivo."""
    return cantidad > 0

def validar_venta(cantidad, stock_actual):
    """Verifica que la cantidad sea positiva y no supere el stock disponible."""
    return cantidad > 0 and cantidad <= stock_actual
```
